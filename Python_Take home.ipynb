```python
# Load the dataset
df = pd.read_excel(r'c:\Users\HP\Downloads/AQI_Levels_2023.xlsx')
```

```python
# Load the dataset
# Replace missing values with column mean (for numeric columns)
numeric_cols = ["AQI", "PM2.5", "PM10", "NO2", "O3", "CO", "SO2", "Temperature", "Humidity", "Wind Speed", "Population", "Hospital Admissions"]
df[numeric_cols] = df[numeric_cols].apply(lambda x: x.fillna(x.mean()))
```

```python
import pandas as pd
```

```python
# Removing Duplicates
# Check for duplicates
print("Duplicates before cleaning:")
print(df.duplicated().sum())
# Remove duplicates
df = df.drop_duplicates()
```

```python
print(df.duplicated().sum())
```

```python
# 3. Correcting Errors
#  Correcting negative values in AQI (assuming negative AQI is an error)
df['AQI'] = df['AQI'].apply(lambda x: abs(x) if x < 0 else x)

```

```python
# 4. Standardizing Data Formats
# Ensure numeric columns are numeric
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)
```
